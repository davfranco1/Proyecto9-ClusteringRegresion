## Encoding

- Dado que, nos interesa asignar peso a nuestras variables de acuerdo con el beneficio (profit) con el que se relacionan, usaremos el Target Encoding para todas las columnas categóricas.

- Como primer paso definimos un diccionario de encoding, llamamos a la clase y almacenamos tanto el nuevo dataframe como el objeto de encoding.

```python
diccionario_encoding = {
    "onehot": [],
    "dummies": [],
    "ordinal": {},
    "label": [],
    "frequency": [],
    "target": ['Segment', 'Market', 'Category', 'Order Priority'],
    "catboost": []
}
```

```python
encoding = se.Encoding(dataframe=df, diccionario_encoding=diccionario_encoding, variable_respuesta="Profit")
```

```python
ta_enc, encoder = encoding.target_encoding()
ta_enc.head()
```

## Estandarización

- Para mantener la escala de nuestros datos, estandarizaremos tanto la columna numérica ("profit") como el resto de categóricas (a pesar de estar encoded), de modo que, nuestro modelo asigne "pesos" equitativos.

# Clustering

- Para crear nuestros clusters, usaremos en este caso el modelo K-Means. Primero, llamamos a la clase.

```python
clustering = sc.Clustering(df_standard)
```

- Y buscamos el "codo" de nuestros datos, aquel punto donde el silhouette es mayor.

```python
clustering.sacar_clusters_kmeans()
```

- Hemos identificado que nuestro silhouette más alto es con k=2, así que vamos a calcular sus métricas.

```python
df_kmeans, labels, cluster_model= clustering.modelo_kmeans(df_standard, 2)
```

- Buscamos que el silhouette sea lo más cercano a 1, mientras el DBI lo más próximo a 0, con una cardinalidad que refleje una cantidad de elementos que tenga sentido con nuestro conjunto de datos.

- En este caso, vemos que nuestro silhouette es apenas aceptable, estando por encima de 0.5. El silhouette mide qué tan similar es un punto a su cluster. Un valor de 1 indica distancias cortas dentro de su propio cluster y lejanas con otros.

- El DBI, por su parte, explica lo bien que se han separado los clusters. Lo ideal es que tienda a 0. Hasta un valor de 1, indica una separación moderada, con solapamiento en algunos casos. Por encima de 1, se entiende que la separación ha sido pobre. En este caso, un valor de 0.87 nos permite continuar trabajando con estos resultados.

- En lo que respecta a la cardinalidad, tenemos un grupo principal, y otro más pequeño con cerca de 10,000 datos menos.

```python
clustering.calcular_metricas(labels)
```

- Para entender si esta separación tiene algún sentido, vamos a ver dos gráficas, primero un radar plot:

```python
variables = ['Profit_minmax', 'Segment_minmax', 'Market_minmax', 'Category_minmax',
       'Order Priority_minmax']

clustering.radar_plot(df_kmeans, variables, "clusters_kmeans")
```

- Aquí, podemos ver que nuestros clusters se han dividido principalmente por el profit que generan. Para entenderlo mejor, vamos a ver un barplot.

```python
df_kmeans.columns = list(df.columns)+["clusters_kmeans"]
df["clusters_kmeans"] = df_kmeans["clusters_kmeans"]
```

```python
clustering.plot_completo_clusters(data=df, x=df.columns, hue='clusters_kmeans')
```

- En estos barplots, donde tenemos uno para cada feature, podemos ver las siguientes características los clusters generados:

> Cluster 0
- Aglomera las ventas de los segmentos Consumer y Corporate, cuyas ventas se centran en tecnología y muebles.

> Cluster 1
- Cubre exclusivamente al segmento de Home Office, aunque participa también de Consumer y Corporate. Su principal categoría de ventas son los suministros de oficina. Genera beneficios que superan en más de 2x los obtenidos por el cluster 1.
